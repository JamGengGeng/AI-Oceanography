```python
import matplotlib.pyplot as plt
import numpy as np
import skill_metrics as sm

y = np.random.normal(0, 1, 100)
y_hat = np.random.normal(0, 5, 100)


taylor_stats = sm.taylor_statistics(y, y_hat, 'data')

# Store statistics in arrays
sdev = taylor_stats['sdev'][0]
crmsd = taylor_stats['crmsd'][0]
ccoef = taylor_stats['ccoef'][0]

sm.taylor_diagram(sdev,crmsd,ccoef, styleOBS = '-', colOBS = 'r', markerobs = 'o', titleOBS = 'observation')
```
